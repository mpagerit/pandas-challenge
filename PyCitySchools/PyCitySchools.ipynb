```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#find the values specified in instructions
TotalSchools = len(school_data_complete['school_name'].unique())
TotalStudents = len(school_data_complete['student_name'])
AvgMath = school_data_complete['math_score'].mean()
AvgReading = school_data_complete['reading_score'].mean()
TotalBudget = school_data_complete['budget'].unique().sum()

#find the number of students passing reading, then calculate the % passing
passReading = len(school_data_complete['reading_score'].loc[(school_data_complete['reading_score'] >= 70 )])
pctReading = passReading / TotalStudents * 100

#find the number of students passing math, then calculate hte % passing
passMath = len(school_data_complete['math_score'].loc[(school_data_complete['math_score'] >= 70 )])
pctMath = passMath / TotalStudents * 100

#take the number of students passing reading
allPass = school_data_complete.loc[(school_data_complete['reading_score'] >= 70 )]

#filter for students also passing math, then calculate the % passing
passing = len(allPass['math_score'].loc[(allPass['math_score'] >= 70 )])
pctAll = passing / TotalStudents * 100

District_df = pd.DataFrame({'Total Schools':[TotalSchools], 'Total Students':[TotalStudents],
                           'Total Budget':[TotalBudget], "Average Math Score":[AvgMath], 
                           'Average Reading Score':[AvgReading], "% Passing Math":[pctMath],
                           '% Passing Reading':[pctReading], '% Overall Passing':[pctAll]})
#format the data
District_df['Total Students'] = District_df['Total Students'].astype(int).map("{:,}".format)
District_df['Total Budget'] = District_df['Total Budget'].astype(float).map("${:,.2f}".format)
District_df['Average Math Score'] = District_df['Average Math Score'].astype(float).map("{:,.2f}".format)
District_df['Average Reading Score'] = District_df['Average Reading Score'].astype(float).map("{:,.2f}".format)
District_df['% Passing Math'] = District_df['% Passing Math'].astype(float).map("{:,.2f}%".format)
District_df['% Passing Reading'] = District_df['% Passing Reading'].astype(float).map("{:,.2f}%".format)
District_df['% Overall Passing'] = District_df['% Overall Passing'].astype(float).map("{:,.2f}%".format)

District_df
```
